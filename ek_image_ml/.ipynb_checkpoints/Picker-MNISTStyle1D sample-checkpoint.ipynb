```python
# Sklearn scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
# Keras
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.preprocessing import image
```

### Load two PlentyOfFish images

```python
strImgPath0 = '../ek_scrape/img/doctordata/MyCity08mi/HappyAustinGal.00.png'
img0 = pyplot.imread(strImgPath0, format='jpg')
pyplot.imshow(img0)
```

```python
arrX0 = image.img_to_array(img0)
arrX0.shape
```

```python
strImgPath1 = '../ek_scrape/img/doctordata/MyCity08mi/kbriginatx.02.png'
img1 = pyplot.imread(strImgPath1, format='jpg')
pyplot.imshow(img1)
```

```python
# Preprocess images for model prediction
# This step handles scaling and normalization
arrX1 = image.img_to_array(img1)
arrX1.shape
```

```python
arrTrainX = np.array([arrX0, arrX1])
arrTrainX.shape
```

```python
arrTrainY = np.array([[1], [0]])
arrTrainY.shape
```

```python
strImgPath2 = '../ek_scrape/img/doctordata/MyCity08mi/HappyAustinGal.01.png'
img2 = pyplot.imread(strImgPath2, format='jpg')
pyplot.imshow(img2)
```

```python
arrX2 = image.img_to_array(img2)
arrTestX=np.array([arrX2])
arrTestX.shape
```

```python
arrTestY = np.array([[1]])
arrTestY.shape
```

## Scaling and Normalization

We use Sklearn's MinMaxScaler to normalize our data between 0 and 1

```python
# transform the 110x110x3 pics to a flat 1D array
fltDimCount = arrTrainX.shape[1] * arrTrainX.shape[2] * arrTrainX.shape[3]
arrTrainX = arrTrainX.reshape(arrTrainX.shape[0], fltDimCount)
arrTrainX.shape
```

```python
arrTestX = arrTestX.reshape(arrTestX.shape[0], fltDimCount)
arrTestX.shape
```

```python
# Next, we normalize our training data to be between 0 and 1
scaler = MinMaxScaler().fit(arrTrainX)
arrTrainX = scaler.transform(arrTrainX)
arrTestX = scaler.transform(arrTestX)
```

## Our first step is to create an empty sequential model

```python
# Create an empty sequential model
model = Sequential()
```

## Next, we add our first hidden layer

In the first hidden layer, we must also specify the dimension of our input layer. This will simply be the number of elements (pixels) in each image.

```python
# Add the first layer where the input dimensions are the 784 pixel values
# We can also choose our activation function. `relu` is a common
model.add(Dense(100, activation='relu', input_dim=arrTrainX.shape[1]))
```

## We then add a second hidden layer with 100 densely connected nodes

A dense layer is when every node from the previous layer is connected to each node in the current layer.

```python
# Add a second hidden layer
model.add(Dense(100, activation='relu'))
```

## Our final output layer uses a `softmax` activation function for logistic regression.

We also need to specify the number of output classes. In this case, the number of digits that we wish to classify.

```python
# Add our final output layer where the number of nodes 
# corresponds to the number of y labels
model.add(Dense(1, activation='softmax'))
```

## Compile and Train our Model

Now that we have our model architecture defined, we must compile the model using a loss function and optimizer. We can also specify additional training metrics such as accuracy.

```python
# categorical hinge
model.compile(optimizer='adam', 
              loss='categorical_hinge', 
              metrics=['accuracy'])
```

## Finally, we train our model using our training data

Training consists of updating our weights using our optimizer and loss function. In this example, we choose 10 iterations (loops) of training that are called epochs.

We also choose to shuffle our training data and increase the detail printed out during each training cycle.

```python
# Fit (train) the model
model.fit(
    arrTrainX,
    arrTrainY,
    epochs=10,
    shuffle=True,
    verbose=2
)
```

## Evaluating the Model

We use our testing data to validate our model. This is how we determine the validity of our model (i.e. the ability to predict new and previously unseen data points)

```python
# Evaluate the model using the training data 
model_loss, model_accuracy = model.evaluate(arrTestX, arrTestX, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
